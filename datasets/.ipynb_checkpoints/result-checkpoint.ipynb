### 3. Check and fetch mail number from the email body section if any present and store it in another csv.

```python
phone=[]
for i in data['Email Boby']:
    text =i
    regex = re.compile("\(?\d{3}\)?[ -]?\d{3}[-]?\d{4}")
    phone.append(re.findall(regex, text))
```

```python
phone_no=pd.DataFrame(phone)
```

```python
phone_no.to_csv('mail_no.csv')
```

### 2. Check and fetch email address from the email body section if any present and store it in another csv. 

```python
email=[]
for i in data['Email Boby']:
    email.append(re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", i))
```

```python
email
```

```python
email=[]
for i in data['Email Boby']:
    email.append(i.replace(' [at] ','@'))
```

```python
email
```

```python
email1=[]
for i in email:
    email1.append(i.replace('[at]','@'))
```

```python
email1
```

```python
email_body=[]
for i in email1:
    email_body.append(re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+", i))
```

```python
email_body
```

```python
email_id=pd.DataFrame(email_body)
```
